```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from datetime import timedelta
import sys
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ferdinjoe/Jupyter_Data_Visualization/master/kanjanaphisek_bangkok_pm25.csv')
df
```

```python
df.rename(columns={'UTC Hour':'Hour'}, inplace=True)
df.head()
```

```python
df['datetime'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour']]) + pd.Timedelta(hours=7) # UTC to Bangkok time
df.head()
```

```python
df.index = df['datetime']
df.head()
```

```python
df.Hour = df.index.hour
df.head()
```

```python
df.drop(columns=['PM10_mask','Retrospective','datetime'],inplace=True)
df.head()
```

```python
condition_1 = df['PM2.5'] <= 50
condition_2 = df['PM2.5'] >= 51
condition_3 = df['PM2.5'] <= 100
condition_4 = df['PM2.5'] >= 101
Low = df[(condition_1)]
Med = df[(condition_2) & (condition_3)]
High = df[(condition_4)]
```

```python
plt.figure(figsize=(15, 6))
plt.scatter(Low.index, Low['PM2.5'], color = "green",label="Low")
plt.scatter(Med.index, Med['PM2.5'], color = "yellow",label="Med")
plt.scatter(High.index, High['PM2.5'], color = "red",label="High")
plt.legend(numpoints=1)
plt.xlabel('Datetime')
plt.ylabel('PM 2.5')
plt.title('PM 2.5')
```

```python
print(Low['PM2.5'].count())
print(Med['PM2.5'].count())
print(High['PM2.5'].count())
```

```python
sizes = [22006, 1627, 90]
labels = ['Low', 'Med', 'High']
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, shadow=True, autopct='%1.2f%%')
ax.axis('equal')
ax.set_title('PM2.5')
```

```python
explode = (0.1, 0.1, 0.4)
fig, ax = plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='%1.2f%%',
       shadow=True,
       startangle=270)
ax.axis('equal')
ax.set_title('PM2.5')
plt.show()
```
